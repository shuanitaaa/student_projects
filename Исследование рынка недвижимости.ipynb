```python
# визуализация долей пропущенных значений
def pass_value_barh(df):
    (
        (df.isna().mean()*100)
        .to_frame()
        .rename(columns = {0:'space'})
        .query('space > 0')
        .sort_values(by = 'space', ascending = True)
        .plot(kind= 'barh', figsize=(20,10),legend = False, fontsize = 16)
        .set_title('Доли пропущенных значений в исследуемых данных' + "\n", fontsize = 22)    
    );    
```

```python
pass_value_barh(df)
```

Сразу видно, что в названии населенных пунктов присутсвует разный вариант написания слова "поселок" - через "е" и через "ё". Это говорит о том, что, вероятнее всего, данные вносились вручную, а не выбирались из раскрывающегося списка. 

Так же есть написание nan (это те самы пропущенные значения). Помимо этого предположу, что может быть разница в написании с точки зрения регистра. Это тоже сразу поправлю, чтобы в дальнейшем не было затруднений.

### Добавление столбца со стоимостью кв. м.

```python
df['meter_price'] = (df['last_price'] / df['total_area'])
#переввод в формат int чтобы избавиться от знаковв после запятой
df.head()
```

Так как я вычислила стоимость квадратного метра, переведу стоимость квартир в млн., а так же добавлю в датафрейм столбец с категорией цены:
- до 2 млн.
- от 2 до 4 млн.
- от 4 до 6 млн.
- от 6 до 8 млн.
- от 8 до 10 млн.
- от 10 до 15 млн.
- от 15 до 20 млн.
- от 20 до 30 млн.
- от 30 до 50 млн.
- от 50 до 100 млн.
- ужасно дорого.

```python
df['last_price'] = (df['last_price'] / 1000000).round(2) 

def price_tipe(data):
    object_price = data['last_price']
    if object_price < 2:
        return 'до 2 млн.'
    if 2 <= object_price <= 4:
        return 'от 2 до 4 млн.'
    if 4 <= object_price <= 6:
        return 'от 4 до 6 млн.'
    if 6 <= object_price <= 8:
        return 'от 6 до 8 млн.'
    if 8 <= object_price <= 10:
        return 'от 8 до 10 млн.'
    if 10 <= object_price <= 15:
        return 'от 10 до 15 млн.'
    if 15 <= object_price <= 20:
        return 'от 15 до 20 млн.'
    if 20 <= object_price <= 30:
        return 'от 20 до 30 млн.'
    if 30 <= object_price <= 50:
        return 'от 30 до 50 млн.'    
    if 50 <= object_price <= 100:
        return 'от 50 до 100 млн.'     
    
    else:
        return 'ужасно дорого'

df['price_tipe'] = df.apply(price_tipe, axis=1)

df.head(15)
```

Добавлю категорию этажей через функцию.

```python
def floor_tipe(data):
    object_floor = data['floor']
    total_floors = data['floors_total']
    if object_floor == 1:
        return 'первый'
    if object_floor == total_floors:
        return 'последний'

    else:
        return 'другой'

df['floor_tipe'] = df.apply(floor_tipe, axis=1)

df.head(15)
```
